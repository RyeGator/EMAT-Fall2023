# Is there a correlation between the length/duration of the tracks in Spotify and their respective genres?
Does one genre seem to have longer track lengths than another?                    
Do they all seem similer in length?

### Get access to Spotify

```python
import requests
import pandas as pd
import json
import base64
import urllib
import datetime
```

```python
Client_ID = pd.read_csv('spotify_keys.txt')['Client_ID'].iloc[0]
```

```python
Client_Secret = pd.read_csv('spotify_keys.txt')['Client_Secret'].iloc[0]
```

```python
#Client_ID
```

```python
#Client_Secret
```

```python
client_cred = base64.b64encode(str(Client_ID + ":" + Client_Secret).encode('ascii'))
```

```python
#client_cred
```

```python
headers = {"Authorization": "Basic {}".format(client_cred.decode("ascii"))}
```

```python
#headers
```

```python
payload = {"grant_type" : "client_credentials"}
url = "https://accounts.spotify.com/api/token"
```

```python
session_key_responce = requests.post(url = url, data = payload, headers = headers)
```

```python
session_key_responce.status_code
```

```python
#session_key_responce.json()
```

```python
session_header_key = session_key_responce.json()
```

```python
key = session_header_key['access_token']
```

```python
session_headers = {"Authorization": "Bearer {}".format(key)}
```

```python
#session_headers
```

### Spotify session is loaded, now define endpoints
Define which genres I'm using  and why
***
According to https://audiocaptain.com/most-popular-music-genres/ the top ten music genres are Pop, Hip Hop, Electronic Dance Music (EDM) , Rock, Rythm & Blues (R&B), Classical, Jazz, Country, K-Pop, and Metal.

# Pop

```python
query_url = urllib.parse.quote('genre:pop')
album_url = 'https://api.spotify.com/v1/search?q={}&type=track&limit=50'.format(query_url)
responce = requests.get(url = album_url, headers = session_headers)
responce.status_code
```

```python
pop_data = json.loads(responce.text)
```

#### Testing

```python
pop_data['tracks']['items'][0]['artists'][0]['name']
```

```python
pop_data['tracks']['items'][0]['name']
```

```python
pop_data['tracks']['items'][0]['duration_ms']
```

#### Turning Test into a list

```python
pop_artist_list = [x['artists'][0]['name'] for x in pop_data['tracks']['items']]
pop_name_list = [x['name'] for x in pop_data['tracks']['items']]
pop_durationMS_list = [x['duration_ms'] for x in pop_data['tracks']['items']]
pop_duration_list = [str(datetime.timedelta(milliseconds=x)) for x in pop_durationMS_list]
```

#### Merging list into dataframe

```python
pop_df = pd.DataFrame(pop_name_list, columns=['Song Title'])
pop_df['Artist'] = pop_artist_list
pop_df['Duration in Min'] = pop_duration_list
```

```python
pop_avgMS = (sum(pop_durationMS_list))/(50)
pop_avg = str(datetime.timedelta(milliseconds=pop_avgMS))
pop_avg
```

```python
pop_df
```
